# Reading a CSV file into pandas dataframe


#### Pandas is a python library that provides the best data structures for data analysis. Especially if the data is in the form of a CSV file. Data from CSV, Excel, SQL databases or text files is stored in an object called DATAFRAME. You can think of dataframe as an n x n matrix.

```python
import pandas as pd 
```

```python
argon9=pd.read_csv('/Users/gurajas/Downloads/exp2/argon28_sy2.csv')
# argon9 is a dataframe which now has argon9 sensor data
```

```python
argon9.head() 
# head command displays the first few rows of a dataframe. 
# Similarly the tail command displays the last few rows of a dataframe
```

```python
argon9["reg_date"] = pd.to_datetime(argon9["reg_date"])
# to_datetime() method converts the values under "reg_date" column into datetime objects
```

```python
argon9.head()
```

#### Matplotlib is a python library used in generating 2D figures. matplotlib.pyplot is a library of command style functions to make changes to the figures. Such as: shading the figures, creating labels etc.  https://matplotlib.org

```python
import matplotlib.pyplot as plt
%matplotlib inline # to display figures inline
%config InlineBackend.figure_format='svg' # figures are all configured to be in SVG format - Scalable Vector Graphics
```

```python
fig, ax1 = plt.subplots()
fig.set_size_inches(10, 7)


ax1.plot(argon9['reg_date'],argon9['pm25_standard'], color='magenta',linewidth=2,label='PM2.5')
```
